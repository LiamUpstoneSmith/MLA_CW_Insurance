# Introduction


```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import numpy as np
import seaborn as sns
```

# Utils

# EDA

```python
df = pd.read_csv("insurance-adapted.csv")
```

### Data Preprocessing

```python
print("The number of rows and columns", df.shape)

# Checking if there are any empty values
print(df.isnull().sum())

# Creates heatmap of missing values
sns.heatmap(df.isnull(), cbar=False)
```

```python
# Dropping the rows with missing values 
clean_df = df.dropna()

# Checking if there are any empty values
print(clean_df.isnull().sum())

# Creates heatmap of values to show no missing values
sns.heatmap(clean_df.isnull(), cbar=False)
```

```python
# Get x
age = clean_df.age

# Get Y
charges = clean_df.charges

smoker = clean_df.smoker
```

```python
label_encoder = LabelEncoder()

clean_df['smoker_encoded'] = label_encoder.fit_transform(clean_df['smoker'])
clean_df = clean_df.drop(columns=['smoker'])
clean_df.head()
```

```python
#plt.scatter(age, charges, color="red")

categories = np.array(clean_df['smoker_encoded'])

# Smokers are in Green, Non-Smokers are in red
colormap = np.array(['r', 'g'])

plt.scatter(age, charges, c=colormap[categories])

plt.title("Check relation betweeen Age and Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
# Age and Charges, labeled based on region

regions  = df.region

# Map regions to colors
color_map = {'southwest': 'blue', 'southeast' : 'red', 'northwest' : 'yellow', 'northeast' : 'green'}
colors = [color_map[region] for region in regions]

# Create scatter plot
plt.scatter(df.age, df.charges, c=colors, edgecolor='k', s=100)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True, linestyle='--', alpha=0.7)

#add legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='SouthWest'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='SouthEast'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow', markersize=10, label='NorthWest'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='NorthEast')
]
plt.legend(handles=legend_elements, title="Regions")

# Show plot
plt.tight_layout()
plt.show()
```
